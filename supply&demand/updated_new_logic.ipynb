```python
import pandas as pd
from IPython.display import display, clear_output
import ipywidgets as widgets

# Load restaurant data
df = pd.read_csv("restaurants.csv")

# Ensure numeric columns are numeric
df["Rating"] = pd.to_numeric(df["Rating"], errors='coerce')
df["Price"] = pd.to_numeric(df["Price"], errors='coerce')
df["Demand_Quantity"] = pd.to_numeric(df["Demand_Quantity"], errors='coerce')

def show_food_filter_form():
    # --- Widget definitions ---
    type_widget = widgets.ToggleButtons(
        options=['supply', 'demand'],
        description='Type:',
        button_style='info'
    )

    quantity_widget = widgets.IntText(value=1, description='Quantity (kg):')
    
    location_widget = widgets.Dropdown(
        options=['Bangalore', 'Pune', 'Chennai'],
        description='Location:'
    )
    
    item_type_widget = widgets.Dropdown(options=['cb', 'mb'], description='Item Type:')
    
    rating_widget = widgets.Dropdown(options=['1','2','3','4','5'], description='Min Rating:')
    price_widget = widgets.Dropdown(options=['120','300','500','700','1000'], description='Max Price:')
    
    filter_button = widgets.Button(description='🔍 Filter Restaurants', button_style='success')
    export_button = widgets.Button(description='💾 Export to Excel', button_style='primary')

    output = widgets.Output()
    filtered_result = pd.DataFrame()  # Store filtered results for export

    # --- Filtering logic ---
    def filter_data(b):
        nonlocal filtered_result
        with output:
            clear_output()
            min_rating = float(rating_widget.value)
            max_price = float(price_widget.value)
            
            filtered_result = df[
                (df["Location"].str.lower() == location_widget.value.lower()) &
                (df["Item_Type"].str.lower() == item_type_widget.value.lower()) &
                (df["Rating"] >= min_rating) &
                (df["Price"] <= max_price) &
                (df["Demand_Quantity"] >= quantity_widget.value)
            ]
            
            print(f"\n🍛 Results for {type_widget.value.capitalize()} Request in {location_widget.value} 🍛\n")
            if filtered_result.empty:
                print("❌ No restaurants match your criteria.")
            else:
                display(filtered_result[["Shopname", "Demand_Quantity", "Location", "Rating", "Item_Type", "Price"]])

    # --- Export logic ---
    def export_data(b):
        if filtered_result.empty:
            with output:
                print("⚠️ No data to export. Please filter first.")
        else:
            # Save to Excel
            filtered_result.to_excel("Filtered_Restaurants.xlsx", index=False)
            with output:
                print("✅ Filtered results exported to 'Filtered_Restaurants.xlsx'")

    filter_button.on_click(filter_data)
    export_button.on_click(export_data)

    # --- Display form ---
    display(widgets.VBox([
        widgets.HTML("<h2>🍽️ Welcome to the Food Filter System 🍛</h2>"),
        type_widget,
        quantity_widget,
        location_widget,
        item_type_widget,
        rating_widget,
        price_widget,
        widgets.HBox([filter_button, export_button]),
        output
    ]))
```
